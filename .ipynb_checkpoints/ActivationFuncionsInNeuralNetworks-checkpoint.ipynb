## O que é uma função de ativação ?

Também conhecida como função de transferência (Transfer Function), <br>
uma função de ativação é utilizada para obter a saída de um nó de uma rede neural.

## Tipos de Função de Ativação

As funções de ativação podem ser divididas em dois tipos:

1-) Linear Activation Function

2-) Non-linear Activation Functions


## Linear Activation Function

<b>Equação:</b> $ f(x) = x $

<b> Conjunto Imagem: </b> $ ] - \infty, + \infty [ $ 

```python
x = np.linspace(-8, 8, 100) 
z = x
  
plt.figure(figsize=(12,8))
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Linear(X)") 
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xticks(np.arange(-8, 10, 2))
plt.yticks(np.arange(-8, 8, 2))

plt.show()
```

### Desvantagem

Quando se usa uma função de ativação linear ou não se usa uma função de ativação, <br>
a saída de cada nó da rede neural será linear ( Z = WX + b ), portanto não importa <br>
quantas hidden layers sua rede tenha, se o problema for complexo (não linear), sua <br>
rede neural será incapaz de aprender o seu funcionamento.

## Non-linear Activation Functions

São as funções mais utilizadas, sua utilização facilita a generalização <br>
por parte do modelo para problemas complexos (não lineares)

Principais função de ativação não lineares:


## Sigmoid or Logistic Activation Function

<b>Equação:</b> $ f(x) = \frac{1}{1 + \exp{(x)}} $

<b> Conjunto Imagem: </b>  ] 0, 1 \[

```python
x = np.linspace(-8, 8, 100) 
z = 1/(1 + np.exp(-x)) 
  
plt.figure(figsize=(12,8))
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Sigmoid(X)") 
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xticks(np.arange(-8, 10, 2))
plt.yticks(np.arange(0, 1.1, 0.5))

plt.show()
```

### Vantagem

Isso é muito útil quando o nosso modelo precisa predizer a probabilidade <br> 
de um certo fenômeno, por exemplo, se uma determinada imagem é uma imagem <br>
de um gato ou não, nesse caso se a saída for maior do que 0.5 o modelo predisse<br>
um sim, caso contrário um não. 


### Desvantagem

1-) Lento para valores muito grandes ou muito pequenos de x.

Quando o valor de z é muito grande ou muito pequeno, a derivada da função sigmoid <br>
naquele ponto é muito próximo de 0, e esse fator pode tornar o aprendizado do modelo <br>
muito lento. 

Essa lentidão se deve ao fato das redes neurais utilizarem o método do <br>
Gradiente Descendente para obter o menor erro para o modelo, e esse método se utiliza <br>
das derivadas para se aproximar do ponto de erro mínimo, logo com uma derivada próxima de 0 <br>
o método levaria muito mais tempo para se aproximar do ponto mínimo.


## Tangent Function or Hyperbolic Tangent Function

<b>Equação:</b> $ f(x) = \frac{\exp{x} - \exp{(-x)}}{\exp{x} + \exp{(-x)}} $

<b> Conjunto Imagem: </b>  ] -1, 1 \[

```python
x = np.linspace(-8, 8, 100) 
z = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
  
plt.figure(figsize=(12,8))
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Hyperbolic Tangent(X)") 
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xticks(np.arange(-8, 10, 2))
plt.yticks(np.arange(-1, 1.1, 0.5))


plt.show()
```

### Vantagem

É notável que a função tangente hiperbólica apresenta derivada maior do <br>
que a função sigmoid em diversos pontos do seu domínio, isso garante um aprendizado <br>
mais rápido do que a função sigmoid.


### Desvantagem

A função Tanh não é utilizada na camada final, pois seu range no eixo y é 2.

Assim como a função sigmoid para valores muito grandes ou muito pequenos a derivada desta <br>
função é próxima de 0, o que retarda muito o aprendizado do modelo.


## Rectified Linear Unit (ReLu)

A preferida da galera para as hidden layers.


<b>Equação:</b> $ f(x) = max(0, x) $

<b> Conjunto Imagem: </b>  [ 0, +$\infty$ \[

```python
x = np.linspace(-8, 8, 100) 
func = lambda x: 0 if x <=0 else x
z = np.array( list( map(func, list(x)) ) )

plt.figure(figsize=(12,8))
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Relu(X)") 
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xticks(np.arange(-8, 10, 2))
plt.yticks(np.arange(0, 10, 2))


plt.show()
```

### Vantagem

Para todo valor de z positivo, a derivada da função nesse ponto é sempre 1, <br>
Logo para valores grandes de z, a função não possui uma derivada próxima de 0 <br>
como nas funções sigmóide e tangente hiperbólica.


### Desvantagens

A função não mapeia os valores negativos de z de maneira apropriada, como é visível no <br>
gráfico, todo valor negativo de z é transformado em 0 pela função.

Não aconselhável de ser utilizada na última camada, pois não restringe a saída entre valores <br>
no intervalo 0 e 1 como a função sigmóide. 


## Por que a função Relu funciona ?

$g(x) = Relu(x) + Relu(x-2) + Relu(x-4) + Relu(x-6) $ 

```python
x = np.linspace(-4, 8, 100) 

func1 = lambda x: 0 if x <= 0 else x

z1 = np.array( list( map(func1, list(x)) ) )
z2 = np.array( list( map(func1, list(x-2)) ) )
z3 = np.array( list( map(func1, list(x-4)) ) )
z4 = np.array( list( map(func1, list(x-6))) ) 


z = z1 + z2 + z3 + z4 

plt.figure(figsize=(15,10))

plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Relu(X)") 
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xticks(np.arange(-4, 10, 2))
plt.yticks(np.arange(0, 20, 2))

plt.show();
```
